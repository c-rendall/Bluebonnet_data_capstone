Read in a csv containing the state abbreviations

```python
state = pd.read_csv('state-abbrevs.csv')
```

Clean the data and add new columns:
1. Use a regular expression to extract the congressional district number
2. Fill (at-large) districts with 0
3. Extract the state from the geographic name column using regular expressions
4. Strip the extra space out of the state column

```python
economic_draft['district'] = economic_draft.name.str.extract('(\d+)(....\d)')[0]
economic_draft['district']= economic_draft.district.fillna('0')
economic_draft['state'] = economic_draft.name.str.extract('[^,]*,(.*)')[0]
economic_draft['state'] = [x.strip() for x in economic_draft['state']]
```

Merge the economic dataset with the state abbreviations dataset

```python
economic_draft = pd.merge(left=economic_draft, right=state, how='inner', on = 'state')
economic_draft['state_dist'] = economic_draft['abbreviation'].astype(str) + '-'+ economic_draft['district'].astype(str)
```
