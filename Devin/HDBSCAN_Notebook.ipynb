```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
import networkx as nx
import seaborn as sns
import hdbscan
```

```python
data = pd.read_csv("cap_data_final3.csv")
data = data.loc[(data['year'] == 2014),:]
data = data.dropna().reset_index()
data = data.drop(columns= ['district', 'state', 'abbreviation', 'year', 'state_dist', 'index'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_rescaled = scaler.fit_transform(data.drop('party', 1))
```

### 2014 Clustering

```python
pca = PCA(0.95)
pca.fit(data_rescaled)
train = pca.transform(data_rescaled)
```

```python
clusterer = hdbscan.HDBSCAN(min_cluster_size=3)
clusterer.fit(train)
clusterer.labels_
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(train[:,0], train[:,1], clusterer.labels_, palette = 'Set2', style = data.party, s =100)
```

```python
clusterer.condensed_tree_.plot()
```
