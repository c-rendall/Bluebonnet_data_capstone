```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.cluster import SpectralClustering, KMeans
from sklearn.metrics import pairwise_distances
from matplotlib import pyplot as plt
import networkx as nx
import seaborn as sns
sns.set()
```

```python
#Load in data
data = pd.read_csv("limiteddata1.csv")
```

```python
#Set your year here: (2014, 2016, or 2018) and clean data
data = data.loc[(data['year'] == 2014),:]
data = data.drop('Unnamed: 0', axis = 1)
data = data.dropna().reset_index()
data = data.drop(columns= ['district', 'abbreviation', 'index'])
```

```python
#Standardize the Data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_rescaled = scaler.fit_transform(data.drop('party', 1))
```

```python
#visualize PCA explained variance
pca = PCA().fit(data_rescaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
#find number of PCs
from sklearn.decomposition import PCA
pca = PCA(0.95)
pca.fit(data_rescaled)
train = pca.transform(data_rescaled)
pca.n_components_
```

```python
#run and plot cluster model
sns.set(rc={'figure.figsize':(11.7,8.27)})
sc = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', random_state=0)
sc_clustering = sc.fit(train)
sns.scatterplot(train[:,0], train[:,1], sc_clustering.labels_, palette = 'Set2', style = data.party, s =100)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sc = SpectralClustering(n_clusters=3, affinity='nearest_neighbors', random_state=0)
sc_clustering = sc.fit(train)
sns.scatterplot(train[:,0], train[:,1], sc_clustering.labels_, palette = 'Set2', style = data.party, s =100)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sc = SpectralClustering(n_clusters=4, affinity='nearest_neighbors', random_state=0)
sc_clustering = sc.fit(train)
sns.scatterplot(train[:,0], train[:,1], sc_clustering.labels_, palette = 'Set2', style = data.party, s =100)
```
